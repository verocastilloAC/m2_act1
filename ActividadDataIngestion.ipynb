### Actividad Data Ingestion

```python
# Importar librer√≠as necesarias de python
import pandas as pd
import requests
import json
import pickle
```

#### 1. Cargar archivos csv como DataFrames

```python
# Archivo 1
path_csv1 = "files/Machine-readable-business-employment-data-sep-2021-quarter.csv"
load_csv1 = pd.read_csv(path_csv1)

load_csv1.head()
```

```python
# Archivo 2
path_csv2 = "files/annual-enterprise-survey-2020-financial-year-provisional-csv.csv"
load_csv2 = pd.read_csv(path_csv2)

load_csv2.head()
```

```python
# Archivo 3
path_csv3 = "files/electronic-card-transactions-november-2021-csv-tables.csv"
load_csv3 = pd.read_csv(path_csv3)

load_csv3.head()
```

#### 2. Cargar archivos JSON como DataFrames

```python
# JSON 1
# Cargar
url1 = "https://jsonplaceholder.typicode.com/todos"
response1 = requests.get(url1)
json_data1 = json.loads(response1.text)
# Convertir
json_df1 = pd.DataFrame(json_data1)

json_df1.head()
```

```python
# JSON 2
# Cargar
url2 = "https://jsonplaceholder.typicode.com/photos"
response2 = requests.get(url2)
json_data2 = json.loads(response2.text)
# Convertir
json_df2 = pd.DataFrame(json_data2)

json_df2.head()
```

```python
# JSON 3
# Cargar
url3 = "https://jsonplaceholder.typicode.com/albums"
response3 = requests.get(url3)
json_data3 = json.loads(response3.text)
# Convertir
json_df3 = pd.DataFrame(json_data3)

json_df3.head()
```

#### 3. Convertir cada DataFrame a .csv, .xlsx y .json

```python
# CSV 1
load_csv1.to_csv("outputs/load_csv1.csv",index=False)
load_csv1.to_excel("outputs/load_csv1.xlsx",index=False,engine='xlsxwriter')
load_csv1.to_json("outputs/load_csv1.json",orient='records')
```

```python
# CSV 2
load_csv2.to_csv("outputs/load_csv2.csv",index=False)
load_csv2.to_excel("outputs/load_csv2.xlsx",index=False,engine='xlsxwriter')
load_csv2.to_json("outputs/load_csv2.json",orient='records')
```

```python
# CSV 3
load_csv3.to_csv("outputs/load_csv3.csv",index=False)
load_csv3.to_excel("outputs/load_csv3.xlsx",index=False,engine='xlsxwriter')
load_csv3.to_json("outputs/load_csv3.json",orient='records')
```

```python
# JSON 1
json_df1.to_csv("outputs/json_df1.csv",index=False)
json_df1.to_excel("outputs/json_df1.xlsx",index=False,engine='xlsxwriter')
json_df1.to_json("outputs/json_df1.json",orient='records')
```

```python
# JSON 2
json_df2.to_csv("outputs/json_df2.csv",index=False)
json_df2.to_excel("outputs/json_df2.xlsx",index=False,engine='xlsxwriter')
json_df2.to_json("outputs/json_df2.json",orient='records')
```

```python
# JSON 3
json_df3.to_csv("outputs/json_df3.csv",index=False)
json_df3.to_excel("outputs/json_df3.xlsx",index=False,engine='xlsxwriter')
json_df3.to_json("outputs/json_df3.json",orient='records')
```

#### 4. Pickle y Un-pickle cada DataFrame

```python
# Convertir a Pickle todos los dataframes
with open('pickles/load_csv1.pickle','wb') as file:
    pickle.dump(load_csv1, file)
with open('pickles/load_csv2.pickle','wb') as file:
    pickle.dump(load_csv2, file)
with open('pickles/load_csv3.pickle','wb') as file:
    pickle.dump(load_csv3, file)
with open('pickles/json_df1.pickle','wb') as file:
    pickle.dump(json_df1, file)
with open('pickles/json_df2.pickle','wb') as file:
    pickle.dump(json_df2, file)
with open('pickles/json_df3.pickle','wb') as file:
    pickle.dump(json_df3, file)
```

```python
# Abrir archivos de Pickle como DF
with open('pickles/load_csv1.pickle','rb') as file:
    df1 = pickle.load(file)
with open('pickles/load_csv2.pickle','rb') as file:
    df2 = pickle.load(file)
with open('pickles/load_csv3.pickle','rb') as file:
    df3 = pickle.load(file)
with open('pickles/json_df1.pickle','rb') as file:
    df4 = pickle.load(file)
with open('pickles/json_df2.pickle','rb') as file:
    df5 = pickle.load(file)
with open('pickles/json_df3.pickle','rb') as file:
    df6 = pickle.load(file)
```
